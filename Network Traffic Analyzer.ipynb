```python
# Step 2.1: Import libraries
import pandas as pd

# Step 2.2: Load Wireshark CSV
df = pd.read_csv("Network_Data.csv")

# Step 2.3: Quick look
print(df.head())
print(df.info())

# Step 2.4: Top source IPs
top_src = df['Source'].value_counts().head(5)
print("Top Source IPs:\n", top_src)

# Step 2.5: Top destination IPs
top_dst = df['Destination'].value_counts().head(5)
print("\nTop Destination IPs:\n", top_dst)

# Step 2.6: Protocol distribution
proto_count = df['Protocol'].value_counts()
print("\nProtocol Distribution:\n", proto_count)
```

```python
import matplotlib.pyplot as plt

# Top source IPs
top_src.plot(kind='bar', title='Top Source IPs')
plt.show()

# Protocol distribution
proto_count.plot(kind='pie', autopct='%1.1f%%', title='Protocol Distribution')
plt.show()
```
